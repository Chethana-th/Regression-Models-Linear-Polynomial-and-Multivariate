```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv('Ice_cream selling data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#Investigate the shape of the data points

X=df[['Temperature (째C)']]
y=df[['Ice Cream Sales (units)']]
plt.scatter(X,y)
plt.xlabel('Temperature (C)')
plt.ylabel('Ice cream sales')
plt.title('Ice cream sales vs Temperature')
plt.show()
```

The relationship in between Temperature and Ice cream sales is non-linear. Therefore Polynomial regression model is suitable for these type of data.

```python
#Fitting the model

poly=PolynomialFeatures(degree=4)
X_poly=poly.fit_transform(X)

model=LinearRegression()
model.fit(X_poly,y)

y_pred=model.predict(X_poly)
```

```python
#plotting the Polynomial curve

plt.scatter(X,y,color='green', label= 'Actual data')
plt.plot(X,y_pred,color='red',label='Predicted data')
plt.xlabel('Temperature')
plt.ylabel('Ice cream sales')
plt.title('Polynomial Regression- Temperature vs Ice cream sales')
```
